```python
import numpy as np
import matplotlib.pyplot as plt

import sys, os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'lib')))
from renormalization import find_Jc, __which_geometry__

print("Cell geometry in use:", __which_geometry__)
```

```python
# Critical point

Jc = find_Jc(
    a=2.1,
    Jlow=1e-2, Jhigh=1e2,
    max_steps=6, max_dist_final=6,
    tol=1e-5,
    growth_threshold=1e4, decay_threshold=1e-4
)
print("Critical Jc =", Jc)
print("Critical Tc =", 1/Jc)
```

```python
# Critical line

a_vals = np.linspace(.8, 4, 10)
Jc_vals = []
Tc_vals = []

for a in a_vals:
    Jc = find_Jc(
        a=a,
        Jlow=1e-3, Jhigh=1e3,
        max_steps=8, max_dist_final=20,
        tol=1e-5,
        growth_threshold=1e3, decay_threshold=1e-3
    )
    Jc_vals.append(Jc)
    Tc_vals.append(1.0 / Jc)
    print(f"a = {a:.4f}  →  Jc = {Jc:.3}, Tc = {1/Jc:.3}")

Jc_vals = np.array(Jc_vals)
Tc_vals = np.array(Tc_vals)

plt.figure(figsize=(7,5))
plt.plot(a_vals, Tc_vals, 'o-', ms=5)
plt.ylim(0, 10)
plt.xlabel("a", fontsize=13)
plt.ylabel("1 / Jc (Critical Temperature)", fontsize=13)
plt.grid(False)
plt.tight_layout()
plt.show()
```
